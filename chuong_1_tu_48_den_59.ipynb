```python
from sympy import *
def C(n,r):
    A = factorial(n)/factorial(r)/factorial(n-r)
    return A
```

```python
# 1.57 1.58
def sols(n,r):
    if n==1:
        return [[r]]
    A=[]
    for k in range (r+1):
        for a in sols(n-1,r-k):
            a.insert(0,k)
            A.append(a)
    return A
```

```python
sols(5,3)
```

```python
def ss(n,r):
    if n==1:
        return [[n]]
    if r==0:
        return [[0 for _ in range (n)]]
    A=[]
    for a in ss(n-1,r):
        a.insert(0,0)
        A.append(a)
    for a in ss(n,r-1):
        a[0]+=1
        A.append(a)
    return A
```

```python
ss(5,3)
```
