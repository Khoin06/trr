```python
# Liệt kê xâu nhị phân
def xaunhiphan(n):
    if n==1:
        return [[0],[1]]
    A = []
    for a in xaunhiphan(n-1):
        for i in range (0,2):
            b=a.copy()
            b.append(i)
            A.append(b)
    return A
xaunhiphan(3)
```

```python
# Liệt kê hoán vị
from sympy import *
def hoanvi(l):
    if len(l)==1:
        return [l]
    A=[]
    n=len(l)
    for i in range (n):
        lc=l.copy()
        x=lc.pop(i)
        for a in hoanvi(lc):
            a.insert(0,x)
            A.append(a)
    return A

hoanvi(['F','I','T'])
```

```python
# Liệt kê chỉnh hợp 
def chinhhop(l,r):
    if r==1:
        return [[i] for i in l]
    A=[]
    for a in chinhhop(l,r-1):
        C= set([i for i in l])-set(a)
        for i in C:
            b=a.copy()
            b.append(i)
            A.append(b)
    return A

chinhhop(['m','r','a','f','t'],3)
```

```python
# Liệt kê hoán vị của các số 1,2,...,n
def hoanviso(n):
    if n==1:
        return [[1]]
    A=[]
    for a in hoanviso(n-1):
        for i in range (n):
            b=a.copy()
            b.insert(i,n)
            A.append(b)
    return A
hoanviso(3)
```

```python
# Liệt kê chỉnh hợp chập r của 1,2,...,n
def chinhhopso(n,r):
    if r==1:
        return [[i] for i in range (1,n+1)]
    A=[]
    for a in chinhhopso(n,r-1):
        C= set([i for i in range (1,n+1)])-set(a)
        for i in C:
            b=a.copy()
            b.append(i)
            A.append(b)
    return A
chinhhopso(4,3)
```

```python
# Liệt kê tổ hợp
def tohop(a,r):
    if r==1:
        return [[i] for i in a]
    A=[]
    n=len(a)
    for i in range (n-r+1):
        for c in tohop(a[i+1:],r-1):
            c=[a[i]]+c
            A.append(c)
    return A
tohop(['1','2','3','4','5','6','7','8','9'],3)
```

```python
# Liệt kê hoán vị lặp
def hoanvilap(l):
    if len(l)==1:
        return [l]
    A=[]
    n=len(l)
    for i in range(n):
        lc=l.copy()
        x=lc.pop(i)
        for a in hoanvilap(lc):
            a.insert(0,x)
            A.append(a)
    B=[]
    for c in A:
        count=0
        for d in B:
            if c==d:
               count +=1 
        if count == 0:
            B.append(c)
    return B
hoanvilap(['c','o','o','l'])
```

```python
# tìm đường đi trong Oxy
def timduongdi(a,b,x,y):
    i=x-a
    j=y-b
    A=[]
    for _ in range(0,i):
        A.append('sang')
    for _ in range (0,j):
        A.append('len')
    return hoanvi(A)
timduongdi(0,0,3,2)
```

```python
# tìm đường đi trong Oxy
def timduongdi2(a,b,c,x,y,z):
    i=x-a
    j=y-b
    k=z-c
    A=[]
    for _ in range(0,i):
        A.append('H')
    for _ in range (0,j):
        A.append('V')
    for _ in range (0,k):
        A.append('A')
    return hoanvi(A)
timduongdi2(-1,2,0,1,3,7)
```

```python
# Liệt kê nghiệm nguyên không âm của ptrinh x1+x2+...+xn=r
# Liệt kê tổ hợp lặp
def timnghiem(n,r):
    if n==1:
        return [[r]]
    A=[]
    for k in range (r+1):
        for a in timnghiem(n-1,r-k):
            a.insert(0,k)
            A.append(a)
    return A
timnghiem(3,10)
```

```python
# liệt kê tổng riêng của số nguyên dương n
def tongrieng(n):
    if n==1:
        return [[1]]
    A=[]
    for k in tongrieng(n-1):
        for a in range (0,len(k)):
            c=k.copy()
            c[a]+=1
            A.append(c)
        k.append(1)
        A.append(k)
    B=[]
    for f in A:
        count=0
        for g in B:
            if f == g:
                count+=1
        if count == 0:
            B.append(f)
    return B
tongrieng(10)
```
