```python
from sympy import *
def C(n,r):
    if r==0 or r==n:
        return 1
    return C(n-1,r)+C(n-1,r-1)
```

```python
def hoanvi(l):
    if len(l)==1:
        return [l]
    A=[]
    n=len(l)
    for i in range(n):
        lc=l.copy()
        x=lc.pop(i)
        for a in hoanvi(lc):
            a.insert(0,x)
            A.append(a)
    B=[]
    for c in A:
        count=0
        for d in B:
            if c==d:
               count +=1 
        if count == 0:
            B.append(c)
    return B
```

```python
# 1.35
# print(C(4,1)*C(3,2)*C(1,1))
print(hoanvi(['c','o','o','l']))
print(len(hoanvi(['c','o','o','l'])))
```

```python
# 1.42
def timduongdi(a,x,b,y):
    i=x-a
    j=y-b
    A=[]
    for _ in range(0,i):
        A.append('sang')
    for _ in range (0,j):
        A.append('len')
    return hoanvi(A)
```

```python
timduongdi(0,3,0,2)
```

```python
# 1.43 1.44
def timduongdi2(a,x,b,y,c,z):
    i=x-a
    j=y-b
    k=z-c
    A=[]
    for _ in range(0,i):
        A.append('H')
    for _ in range (0,j):
        A.append('V')
    for _ in range (0,k):
        A.append('A')
    return hoanvi(A)
# C(len(A),i)*C(len(A)-i,j)*C(len(A)-i-j,k)
```

```python
timduongdi2(-1,1,2,3,0,7)
```
